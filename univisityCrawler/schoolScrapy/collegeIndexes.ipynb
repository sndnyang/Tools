```python
for e in college_set:
    if e['name'].find('(') == -1:
        t = e['name'].replace(',', ' ').replace('-', ' ')

        w = t.split()
        if len(w) == 2:
             
            if w[1] == 'University':
                abbr = w[0]
            elif w[1] == 'College':
                abbr = w[0] + ' ' + w[1][0]
            elif w[0] in ['University', 'College', 'Universidad']:
                abbr = w[0][0] + ' ' + w[1]
            else:
                abbr = t
        else:
            if t.startswith('University of ') and len(w) == 3:
                abbr = w[2]
                #print '%8s  %s' % (abbr, e['name'])
            else: 
                abbr = ''.join(e[0] for e in t.split() if e.lower() not in ['at', '-', 'of', 'the', 'in', 'and', 'de'] and not e.isupper())
                if len(abbr) > 4:
                    print '%8s  %s' % (abbr, e['name'])
        e['name'] += '(%s)' % abbr
```

```python

for u in college_set:
    if os.path.isfile(u['name'] + '.txt'):
        #print(u['name'])
        continue
    url = u['webpage']
    if url[:4] != 'http':
        url = 'http://' + url
    print(url)
    try:
        conn = req.get(url)
        conn.raise_for_status()
        fp = open(u['name'] + ".txt", 'w',encoding='utf-8')
        fp.write(conn.text)
        conn.close()
        fp.close()
    except:
        print('fail   ' + url)
```
