```python

for u in college_set:
    if os.path.isfile(u['name'] + '.txt'):
        #print(u['name'])
        continue
    url = u['webpage']
    if url[:4] != 'http':
        url = 'http://' + url
    print(url)
    try:
        conn = req.get(url)
        conn.raise_for_status()
        fp = open(u['name'] + ".txt", 'w',encoding='utf-8')
        fp.write(conn.text)
        conn.close()
        fp.close()
    except:
        print('fail   ' + url)
```
