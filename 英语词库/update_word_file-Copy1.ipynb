```python
data = xlrd.open_workbook('./gloss.xls')
table = data.sheets()[0]
nrows = table.nrows # 获取表的行数
```

```python
l = ['word', 'level', 'lenovo', 'etyma', 'meanZh', 'meanEn', 'example']

content = []

for i in range(2, nrows): # 循环逐行打印
    if i == 0: # 跳过第一行
        continue
    row = table.row_values(i)[:len(l)]
    if len(row) > 7:
        print len(row), row
        sys.exit()
    content.append(row)
```
